## <b><u>TOKENIZING:</u></b>

```python
df = pd.DataFrame()

# Tokenizing sentences
df["Sentences"] = nltk.sent_tokenize(paragraph)

# Tokenizing words using nltk tokenizer (NOT recommended, as it accepts all punctuations as words too, all it does is splitting by spaces)
# df["Words"] = [nltk.word_tokenize(sentence) for sentence in df["Sentences"]]

# Manual word tokenizing
def word_tokenize(sentence):
    s = copy.deepcopy(sentence)
    # 1. REMOVE PUNCTUATIONS (otherwise they would also be considered as words)
    s = re.sub('[^a-zA-Z0-9\s]', '', s)
    # 2. Convert to lower case
    #    Necessity: To avoid redundant words just due to difference in case, while forming vectors out of words)
    #    For example: "In" and "in" are actually same but would be considered different
    #    Caution: Some words need to be treated different indeed.
    #    Like: "U.S." and "us", both are different words, but after removing punctuation and doing lower case, both map to the same word "us", which is wrong
    s = s.lower()
    s = s.split()
    return s

df["Words"] = [word_tokenize(sentence) for sentence in df["Sentences"]]

df.head()
```

## <b><u>REMOVING STOPWORDS:</u></b>

```python
# Stopwords: Words which are very commonly used in any sentence in a language and that kind of usage of that word is specific to the language, are stop words. 
# Their usage is language specific so they are unimportant and sometimes not much info can be gathered from them about the meaning of the sentence.
# P.S.: Exclude only if unimportant, sometimes may turn out to be important for some usecases. Understand the usecase and infer whether stopwords are important or not...

row_count = len(df.axes[0])
df["Stopword Removed"] = copy.deepcopy(df["Words"].tolist())

for i in range(row_count):
    df["Stopword Removed"][i] = [word for word in df["Stopword Removed"][i] if word not in set(nltk.corpus.stopwords.words('english')) ]

df.head()
```

## <b><u>STEMMING:</u></b>

```python
# Stemming: Replacing all usages of original word with meaningful or non-meaningful stem word
# When to use: When no meaningful output is to be generated, just analysis has to be done; like in sentiment analysis

stemmer = nltk.stem.PorterStemmer()
row_count = len(df.axes[0])
df["Stemmed"] = copy.deepcopy(df["Stopword Removed"].tolist())

for i in range(row_count):
    df["Stemmed"][i] = [stemmer.stem(word) for word in df["Stemmed"][i] ]

df.head()
```

## <b><u>LEMMATIZING:</u></b>

```python
# Lemmatizing: Replacing all usages of original word with meaningful stem word
# When to use: Where a meaningful output has to be returned to user. Like for chatbots, a meaningfully formed sentence is to be returned.

lemmatizer = nltk.stem.WordNetLemmatizer()
row_count = len(df.axes[0])
df["Lemmatized"] = copy.deepcopy(df["Stopword Removed"].tolist())

for i in range(row_count):
    df["Lemmatized"][i] = [lemmatizer.lemmatize(word) for word in df["Lemmatized"][i] ]

df.head()
```
